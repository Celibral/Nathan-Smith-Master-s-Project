```python
# make a list of the variables that contain missing values
vars_with_na = ['Height','Starting_Mass','Mass_1','Mass_2','Mass_3','Mass_4','Starting_SBP','SBP_1','SBP_4','Starting_DBP','DBP_1','DBP_4','Starting_Week', 'Week_1', 'Week_2', 'Week_3','Week_4']
```
